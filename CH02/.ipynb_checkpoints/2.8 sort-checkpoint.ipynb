```python
import numpy as np
#Numpy 快速排序 np.sort(), and np.argsort()

x = np.random.randint(100, size = 100)
print(x)
print("\n")
print(np.sort(x))
print("\n",x,"\n")

x.sort()
print("\n", x)

y = np.random.randint(999, size = 100)
print("\n",y, "\n")

print("\n","sorted index", "\n")
print(np.argsort(y))

i = np.argsort(y)

print("\n",y[i],"\n")
```

```python
# partition 
x = np.array([10,9,3,2,8,3,1,-10,130,-9])

print(np.partition(x,3)) # smallest 3 elements will be parted out, in random order
```
