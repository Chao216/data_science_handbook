```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set plot style

plt.style.use("classic")
x= np.linspace(0.1,np.pi*2,10)
print(x)
```

```python
plt.plot(x)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
x=np.arange(0.1,10,0.01)
figure = plt.figure()
plt.plot(x,np.sin(x))
plt.plot(x,(x**3-x**2+3/x),"--")
plt.show()
figure.savefig("sin.png")
```
