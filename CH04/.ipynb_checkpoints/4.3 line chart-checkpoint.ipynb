```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

```

```python
plt.style.use("seaborn-whitegrid")
```

```python
fig = plt.figure()
ax = plt.axes()

x= np.linspace(0,10,100)
ax.plot(x,x**2)
ax.plot(x,np.cos(x))
ax.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
# line color and style
plt.plot(x,np.sin(x-1), color="green")
plt.plot(x,np.sin(x-2), color="r") # r short for red
plt.plot(x,np.sin(x-3), color="0.69") # grayscale, 1 is white, 0 is black
plt.plot(x,np.sin(x-4), color="#c96c3a")
plt.plot(x,np.sin(x-5), color=(0.2,0.3,0.5)) # only between 0 and 1
```

```python
# linestyle

plt.plot(x,x,linestyle="-") #solid
plt.plot(x,x+0.5,linestyle="--")#dashed
plt.plot(x,x+1,linestyle="-.")#dashdot
plt.plot(x,x+1.5,linestyle=":")#dotted
```

```python
# combine color and linestyle

plt.plot(x,x**3-2*x**2-3*x-1+1/x,"-.r")
```

```python
# adjust x, y axes limits

plt.plot(x,np.sin(x))
plt.xlim(0,30)
plt.ylim(-2,2)
```

```python
plt.plot(x,np.sin(x))
plt.xlim(50,30)
plt.ylim(2,-2)
```

```python
plt.plot(x,x**3-1.67*x)
plt.axis((0,10,-10,10)) # xmin, xmax, ymin, ymax in order
plt.axis("tight") # auto fit the scrren
#plt.axis("equal")
```

```python
# title and axes labels

plt.plot(x,x**2+2*x+1)
plt.title("function in graph")
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
plt.plot(x,x+1,"-b",label="f(x)=x+1")
plt.plot(x,x+2,"-.g",label="f(x)=x+2")
plt.axis((0,10,-10,10))
plt.legend() # to show lable of lines, put ,plt.legend()
```
