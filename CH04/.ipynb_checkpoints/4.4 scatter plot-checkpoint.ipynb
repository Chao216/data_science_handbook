```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn-whitegrid")
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,"o", color="black")
plt.axis("tight")
```

```python
# combine line and scatter
plt.plot(x,y,"-o", color="red")
```

```python
plt.plot(x,y,"-p", color="g",
        markersize=15, linewidth=3,
        markerfacecolor="blue",
        markeredgecolor="black",
        markeredgewidth=3)
```

```python
# use plt.scatter()
plt.scatter(x,y,marker="p")
```

```python
rng = np.random.RandomState(693)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x,y,c=color,s=sizes,alpha=0.3,cmap="viridis")
plt.colorbar()
```

```python
# when dealing with big data, plot is more efficient then scatter
```
