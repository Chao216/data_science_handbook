```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn-whitegrid")
```

```python
data = np.random.randn(1000)
print(data)
```

```python
plt.hist(data)
```

```python
plt.hist(data, bins=30, alpha=0.5, 
 histtype='stepfilled', color='steelblue', 
 edgecolor='red')
```

```python
x1 = np.random.normal(0, 0.8, 1000) 
x2 = np.random.normal(-2, 1, 1000) 
x3 = np.random.normal(3, 2, 1000) 
 
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40) 
 
plt.hist(x1, **kwargs) 
plt.hist(x2, **kwargs) 
plt.hist(x3, **kwargs)
```

```python
# calculate frequency with np.histogram

counts, bin_edges = np.histogram(data, bins=5)
print(counts)
print(bin_edges)
```

```python
counts, boundaries = np.histogram(data, bins=10)
print(counts)
print(boundaries)
print(len(counts))
print(len(boundaries))
```

```python
# 2 dim frequency

mean = [0, 0] 
cov = [[1, 1], [1, 2]] 
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues') 
cb = plt.colorbar() 
cb.set_label('counts in bin')
```

```python
# np.histogram2d

counts, xedges, yedges = np.histogram2d(x,y,bins=10)
print(counts,"\n")
print(xedges,"\n")
print(yedges,"\n")
```

```python
# hexbin

plt.hexbin(x,y,gridsize=20,cmap="Blues") # smaller griszie, bigger hex patterns
cb=plt.colorbar()
cb.set_label("counts in hex")
```
