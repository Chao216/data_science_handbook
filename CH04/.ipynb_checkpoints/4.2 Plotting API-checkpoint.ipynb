```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Matlab style
x= np.arange(1,10,0.01)
plt.figure() #create figure

plt.subplot(2,3,1)
plt.plot(x,np.sin(x))

plt.subplot(2,3,2)
plt.plot(x,np.cos(x))

plt.subplot(2,3,3)
plt.plot(x,np.tan(x))

plt.subplot(2,3,4)
plt.plot(x,x**2)

plt.subplot(2,3,4)
plt.plot(x,1/x)
plt.subplot(2,3,5)
plt.plot(x,2**x)
plt.subplot(2,3,6)
plt.plot(x,np.log(x))

# in plt.subplot(arg1,arg2,arg3) 
# arg1,arg2 shape of figure in matrix form, arg3 number of subplot, 1 to n.
```

```python
# Object-oriented api

fig, ax = plt.subplots(3,3) # shape of figure in matrix row col form
ax[0,0].plot(x,np.sin(x))
ax[0,1].plot(x,np.cos(x))
ax[0,2].plot(x,np.tan(x))
ax[2,0].plot(x,np.log(x))
ax[2,1].plot(x,np.exp(x))
ax[2,2].plot(x,1/x)
fig.savefig("./1.png")
```
