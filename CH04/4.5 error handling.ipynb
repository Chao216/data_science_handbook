```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,50)
dy = 0.8

y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy, fmt=".k")
plt.title("errorbar")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.errorbar(x,y,yerr=dy, fmt="p", color="blue")
plt.title("errorbar")
plt.xlabel("x")
plt.ylabel("y")
```
