```python

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("classic")
```

```python
x = np.linspace(0,5,50)
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,np.sin(x),"-b",label="Sine")
ax[0,1].plot(x,np.cos(x),"--g",label="Cosine")
ax[1,0].plot(x,np.tan(x),":y",label="Tangin")
ax[1,1].plot(x,np.pi**x,"-.r",label="exp_pi")
plt.axis("equal")
ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()
fig.savefig("./Figs/quard.png")
```

```python
x = np.linspace(1,10,100)
plt.plot(x,x**2+3*x-1,"--b",label="y=x^2+3x-1")
plt.plot(x, 2**x, "-.g", label="2^x")
plt.legend(loc="upper center", frameon=True, ncol=2, fancybox = True) # this line will add legends
# loc location, framon , t or f
# ncol matrix shape of legends

# fancybox for round cornor boxes
```
